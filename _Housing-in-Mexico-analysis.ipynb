# Preparing Mexico Data

```python
import pandas as pd
```

```python
# Load CSV files into DataFrames
df1 = pd.read_csv("C:/Users/x/Desktop/Housing-in-Mexico/mexico-real-estate-1.csv",sep=";", index_col=0)
df2 = pd.read_csv("C:/Users/x/Desktop/Housing-in-Mexico/mexico-real-estate-2.csv",sep=";", index_col=0)
df3 = pd.read_csv("C:/Users/x/Desktop/Housing-in-Mexico/mexico-real-estate-3.csv",sep=";", index_col=0)

# Print object type and shape for DataFrames
print("df1 type:", type(df1))
print("df1 shape:", df1.shape)
print()
print("df2 type:", type(df2))
print("df2 shape:", df2.shape)
print()
print("df3 type:", type(df3))
print("df3 shape:", df3.shape)
```

```python
# Print df1 shap
df1.shape
# Print df1 info
df1.info()
# Get output of df1 head
df1.head()
```

```python
# Drop null values from df1
df1.dropna(inplace=True)

# Clean "price_usd" column in df1
df1["price_usd"] = df1["price_usd"].str.replace("$","")
df1["price_usd"] = df1["price_usd"].str.replace(",","").astype(float)

# Print object type, shape, and head
print("df1 type:", type(df1))
print("df1 shape:", df1.shape)
df1.head()
```

## Clean df2

```python
# Print df1 shap
df2.shape
# Print df1 info
df2.info()
# Get output of df1 head
df2.head()
```

```python
# Drop null values from df2
df2.dropna(inplace=True)

# Create "price_usd" column for df2 (19 pesos to the dollar in 2014)
df2["price_usd"] = df2["price_mxn"]/19

# Drop "price_mxn" column from df2
df2.drop(columns="price_mxn",inplace=True)

# Print object type, shape, and head
print("df2 type:", type(df2))
print("df2 shape:", df2.shape)
df2.head()
```

## Clean df3

```python
# Print df1 shap
df3.shape
# Print df1 info
df3.info()
# Get output of df1 head
df3.head()
```

```python
# Drop null values from df3
df3.dropna(inplace=True)
# Create "lat" and "lon" columns for df3
df3[["lat", "lon"]] = df3['lat-lon'].str.split(",",expand=True).astype(float)
# Create "state"  column for df3
df3["state"] = df3["place_with_parent_names"].str.split("|",expand=True)[2]
# drop "place_with_parent_names" and "lat-lon" columns
df3.drop(columns=["place_with_parent_names" , "lat-lon"],inplace=True)
# Print object type, shape, and head
print("df3 type:", type(df3))
print("df3 shape:", df3.shape)
df3.head()
```

## Concatenate DataFrames

```python
# Concatenate df1, df2, and df3
df = pd.concat([df1,df2,df3])

# Print object type, shape, and head
print("df type:", type(df))
print("df shape:", df.shape)
df.head()
```

## Save df

```python
# Save df
df.to_csv("C:/Users/x/Desktop/Housing-in-Mexico/mexico-real-estate-clean.csv",index=False)
```

## Research Question 2 :Is there a relationship between home size and price?

```python
# scatter plot of "price_usd" vs "area_m2"
plt.scatter(df['area_m2'],df['price_usd'])

# Add x-axis label
plt.xlabel("Area [sq meters]")
# Add y-axis label
plt.ylabel("Price [USD]")
# Add title
plt.title("Price VS Area");
```
